# **Linear Regression - Closed Form Solution**

## **Introduction**
Linear Regression is a fundamental supervised learning algorithm used for predicting continuous values. The goal is to find the best-fit line that minimizes the error.

The equation of a simple linear regression model is:

\[
y = w_1 x + w_0
\]

Where:
- \( y \) is the **target variable**.
- \( x \) is the **feature (independent variable)**.
- \( w_1 \) is the **slope (weight)**.
- \( w_0 \) is the **intercept (bias term)**.

### üéØ How to Find \( w_1 \) and \( w_0 \)?

Using the **Least Squares Method**, we minimize the **Mean Squared Error (MSE)**:

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - (w_1 x_i + w_0) \right)^2
$$


## Cost Function ùêΩ(ùë§) and its Derivatives

$$
J(w) = \sum_{i=1}^{n} \left( y^{(i)} - w_0 - w_1 x^{(i)} \right)^2
$$


$$
\frac{\partial J(w)}{\partial w_1} = \sum_{i=1}^{n} 2 \left( y^{(i)} - w_0 - w_1 x^{(i)} \right) (-x^{(i)}) = 0
$$

$$
\frac{\partial J(w)}{\partial w_0} = \sum_{i=1}^{n} 2 \left( y^{(i)} - w_0 - w_1 x^{(i)} \right) (-1) = 0
$$

$$
\left( \sum_{i=1}^{n} y_i - n w_0 - w_1 \sum_{i=1}^{n} x_i \right) = 0
$$

$$
w_0 = \frac{Y - w_1 X}{n}
$$

$$
\left( \sum_{i=1}^{n} x_i y_i - w_0 \sum_{i=1}^{n} x_i - w_1 \sum_{i=1}^{n} x_i^2 \right) = 0
$$

$$
w_1 = \frac{n \sum_{i=1}^{n} x_i y_i - X Y}{n \sum_{i=1}^{n} x_i^2 - X^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data (one feature)
np.random.seed(42)
X = np.linspace(1, 10, 100)  # Feature values
y = 3 * X + 5 + np.random.randn(100) * 3  # True relation with noise

# Number of samples
n = len(X)

# Compute summations
sum_x = np.sum(X)
sum_y = np.sum(y)
sum_xy = np.sum(X * y)
sum_x_squared = np.sum(X ** 2)

# Compute w1 using the detailed formula
w1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)

# Compute w0 using the derived formula
w0 = (sum_y - w1 * sum_x) / n

# Print computed parameters
print(f"Intercept (w0): {w0:.2f}")
print(f"Slope (w1): {w1:.2f}")

# Predict values
y_pred = w1 * X + w0

# Plot the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, y_pred, color='red', label=f"Prediction Line (y = {w1:.2f}x + {w0:.2f})")
plt.xlabel("Feature (X)")
plt.ylabel("Target (Y)")
plt.title("Linear Regression - One Parameter")
plt.legend()
plt.grid(True)
plt.show()
```
